```python
from datetime import datetime
import pandas as pd

# import pandas library
todays_date = datetime.now().strftime("%Y-%m-%d")
student_id = "edwgol5635"

file_path = r'C:\Files\Excel\DA_PythonR\2019_Happiness_Index_v2_Task1.xlsx'

# Load excel file
df = pd.read_excel(file_path)

# Find duplicated rows before cleaning
duplicate_rows_before = df[df.duplicated()]

print(f"### {student_id} – {todays_date}")
print("## Duplicate Rows before cleaning")
print(duplicate_rows_before)

# Make copy of data frame
df_cleaned = df.copy()

# Delete duplicates
df_cleaned.drop_duplicates(inplace=True)

# Find duplicated rows after cleaning
duplicate_rows_after = df_cleaned[df_cleaned.duplicated()]

print("## Duplicate Rows after cleaning")
print(duplicate_rows_after)
```
